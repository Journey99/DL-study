### ultralytics (yolov5, yolov8, yolov11)

! pip install ultralytics

```python
from ultralytics import YOLO

# 예: yolov8n (ultralytics 제공), 또는 공식 문서에 따라 yolov11s 등도 사용 가능
model = YOLO("yolov8n.pt")  
# 이미지 추론
results = model("images/test.jpg", conf=0.25, imgsz=640)
results.print()     # 요약 출력
results.show()      # 시각화

# 비디오/웹캠
model.track(source=0, show=True)  # 웹캠 실시간
```
